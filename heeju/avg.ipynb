```python
import csv
import pandas as pd
import numpy as np

cctv = pd.read_csv('./grid/cctv_clear.csv', encoding='utf-8', dtype={'위도':np.float, '경도':np.float})
light = pd.read_csv('./grid/merge_extract_light.csv', encoding='utf-8')
bell = pd.read_csv('./grid/Seoul_Bell.csv')
```

```python
print(cctv)
print(light)
print(bell)
```

```python
# Data Frame의 위도 경도값을 보고 arr의 알맞는 위치에 개수를 셈

def count(df, c1, c2):
    # 구하려고 하는 범위의 위도 경도 범위
    ub = 37.700393
    bb = 37.426793
    height = ub-bb
    h_n = int(round(height/0.003600,0))
    lb = 126.764095
    rb = 127.182595
    width = rb-lb
    w_n = int(round(width/0.004500, 0))
    # 각 구역의 개수를 저장할 DataFrame 생성. 0으로 초기화
    d = pd.DataFrame(0, index=np.arange(h_n+1), columns=np.arange(w_n+1))

    
    for i in df.index:
        # 구하고자 하는 범위 내에 있으면 진행
        lat = float(df.loc[i, c1])
        lng = float(df.loc[i, c2])
        if bb <= lat and lat <= ub:
            if lb <= lng and lng <= rb: 
                # 몇번째 그리드 내에 있는지 계산
                i1 = int((lat-bb)/0.003600)
                i2 = int((lng-lb)/0.004500)
                x = d.iloc[i1,i2]
                d.iat[i1, i2] = x+1
    return d
                
```

```python
cctv_df = count(cctv, '위도', '경도')
print(cctv_df)
# cctv_df.to_csv("./cctv결과.csv", header=False, index=False)
```

```python
light_df = count(light, 'lat', 'lng')
print(light_df)
# light_df.to_csv("./보안등결과.csv", header=False, index=False)
```

```python
bell_df = count(bell, '위도', '경도')
print(bell_df)
# bell_df.to_csv("./안전벨결과.csv", header=False, index=False)
```
